## 2. Create Point Differential Tag

```python
# first create pvt with necessary cols

pv_pts_dif = df_sbs_finalv2_4.pivot_table(index='Superbowl_#', columns='SUPERBOWL (Pts)', values='SUPERBOWL (Pts)')
```

```python
pv_pts_dif.head()
```

```python
# find max and min value of each row

pv_pts_dif['maxValue'] = pv_pts_dif.max(axis=1)
```

```python
# check new col

pv_pts_dif.head()
```

```python
pv_pts_dif['minValue'] = pv_pts_dif.min(axis=1)
```

```python
# check col

pv_pts_dif.head()
```

```python
# reset index

pv_pts_dif = pv_pts_dif.reset_index()
```

```python
# check result

pv_pts_dif.head()
```

```python
# delete unneeded cols

pv_pts_dif_del_cols = pv_pts_dif[['Superbowl_#', 'maxValue', 'minValue']]
```

```python
# check result

pv_pts_dif_del_cols.head()
```

```python
# find diff of values

pv_pts_dif_del_cols['Pts_differential'] = pv_pts_dif_del_cols['maxValue'] - pv_pts_dif_del_cols['minValue']
```

```python
# check new col

pv_pts_dif_del_cols.head()
```

```python
# delete unneeded cols

pv_pts_dif_fin = pv_pts_dif_del_cols[['Superbowl_#', 'Pts_differential']]
```

```python
# check df

pv_pts_dif_fin.head()
```

```python
# merge and add Pts_differential col

df_sbs_finalv2_5 = df_sbs_finalv2_4.merge(pv_pts_dif_fin, on=['Superbowl_#'], how='left')
```

```python
# check result

df_sbs_finalv2_5.head()
```

>### A. Create 3 categories for pts diff
>>#### 1. Low Pts Diff (=<8)
>>#### 2. Mid Pts Diff (>8 & <17)
>>#### 3. High Pts Diff (=>17)

```python
df_sbs_finalv2_5.loc[df_sbs_finalv2_5['Pts_differential'] <= 8, 'Pts_Diff_Category'] = 'Low Pts Diff'
```

```python
df_sbs_finalv2_5.loc[(df_sbs_finalv2_5['Pts_differential'] > 8) & (df_sbs_finalv2_5['Pts_differential'] <17), 'Pts_Diff_Category'] = 'Mid Pts Diff'
```

```python
df_sbs_finalv2_5.loc[df_sbs_finalv2_5['Pts_differential'] >= 17, 'Pts_Diff_Category'] = 'High Pts Diff'
```

```python
# check results

df_sbs_finalv2_5.head(30)
```

## 4. Create New Total Turnovers column

```python
df_tot_turnovers = df_sbs_final_v4[['Key', 'O_Passing (Int)', 'O_Fumbles (FL)']]
```

```python
# check result

df_tot_turnovers.head()
```

```python
# Group the DataFrame by the 'Key' column

grouped_df = df_tot_turnovers.groupby('Key')

# Apply the agg() function to the grouped DataFrame

agg_tot_turnovers = grouped_df.agg('sum')
```

```python
# check result

agg_tot_turnovers.head(116)
```

```python
# create total turnovers col

agg_tot_turnovers['Total_turnovers'] = agg_tot_turnovers.sum(axis=1)
```

```python
# check col

agg_tot_turnovers.head()
```

```python
# reset index

agg_tot_turnovers = agg_tot_turnovers.reset_index()
```

```python
# remove unnecessary cols

tot_turns = agg_tot_turnovers[['Key', 'Total_turnovers']]
```

```python
# check df

tot_turns.head()
```

```python
# merge and add Total_turnovers col

df_sbs_final_v5 = df_sbs_final_v4.merge(tot_turns, on=['Key'], how='left')
```

```python
# check col

df_sbs_final_v5.head()
```
