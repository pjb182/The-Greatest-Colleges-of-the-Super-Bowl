### Ave age per College per team

```python
# create pivot table

pv_college = df_sbs_v7.pivot_table(index='Player_Stats (College/Univ)', columns='Player_Stats (Key)', values='Player_Stats (Age)', aggfunc='mean')

```

```python
# check pvtable

pv_college.head()
```

```python
# reset index

pv_college = pv_college.reset_index()
```

```python
# check index

pv_college.head()
```

```python
# melt pvt

# Melt the dataframe without losing index
pv_college_melted = pd.melt(pv_college, id_vars=['Player_Stats (College/Univ)'])
```

```python
# check result

pv_college_melted.head()
```

```python
# Rename the column 'value' to 'Univ_Avg_Age_by_Team'

pv_college_melted = pv_college_melted.rename(columns={'value': 'Univ_Avg_Age_by_Team'})
```

```python
# check col name

pv_college_melted.head()
```

```python
# merge pv to v8

df_sbs_v8 = df_sbs_v8.merge(pv_college_melted, on=['Player_Stats (Key)', 'Player_Stats (College/Univ)'], how='outer')
```

```python
# create separate total columns

# Create lsits of the columns to sum
total_yds = ['TEAM_O_Passing (Yds)', 'TEAM_O_Passing (-Yds)', 'TEAM_O_Rushing (Yds)']
total_tds = ['TEAM_O_Passing (TD)', 'TEAM_O_Rushing (TD)', 'TEAM_D_Interceptions (TD)', 'TEAM_SPT_Kick_Returns (TD)', 'TEAM_SPT_Punt_Returns (TD)']
total_turnovers = ['TEAM_O_Passing (Int)', 'TEAM_O_Fumbles (FL)']

# Create new columns based on lists
df['TEAM_total_yds'] = df[total_yds].sum(axis=1)
df['TEAM_total_tds'] = df[total_tds].sum(axis=1)
df['TEAM_total_turnovers'] = df[total_tds].sum(axis=1)
```
